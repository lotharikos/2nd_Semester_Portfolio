```python
import seaborn as sns
import matplotlib.pyplot as plt

# Reload the cleaned dataset
file_path = "/mnt/data/studentdrop out data.csv"
df = pd.read_csv("C:\\Users\\lothar\\OneDrive\\Desktop\\python and data sets\\data set\\studentdrop out data.csv", delimiter=';')
df['Target_encoded'] = df['Target'].map({'Dropout': 0, 'Enrolled': 1, 'Graduate': 2})

# Selected variables for correlation heatmap
top_vars = [
    'Curricular units 2nd sem (approved)',
    'Curricular units 2nd sem (grade)',
    'Curricular units 1st sem (approved)',
    'Curricular units 1st sem (grade)',
    'Admission grade',
    'Target_encoded'
]

# Compute correlation
corr_subset = df[top_vars].corr().round(2)

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_subset, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Top Factors Influencing Graduation")
plt.tight_layout()
plt.show()

corr_subset
```

## 🔍 Key Factors Influencing Graduation (Insights from Correlation)

1. **Curricular units 2nd sem (approved)** — *Correlation: 0.62*
   - Strongest predictor of graduation.
   - Students who passed more 2nd semester subjects were significantly more likely to graduate.

2. **Curricular units 2nd sem (grade)** — *Correlation: 0.57*
   - Higher grades in the second semester strongly relate to graduation.

3. **Curricular units 1st sem (approved)** — *Correlation: 0.53*
   - Students passing more courses early on (first semester) also showed better graduation outcomes.

4. **Curricular units 1st sem (grade)** — *Correlation: 0.49*
   - Good academic start is valuable, but slightly less impactful than second-semester performance.

5. **Admission grade** — *Correlation: 0.12*
   - Weak relationship with graduation.
   - Suggests performance after enrollment matters more than pre-admission scores.


```python
# Set plot style
sns.set(style="whitegrid")

# Plot 1: Boxplot - Curricular units 2nd sem (approved) by Target
plt.figure(figsize=(8, 5))
sns.boxplot(x='Target', y='Curricular units 2nd sem (approved)', data=df, palette='Set2')
plt.title('2nd Semester Approvals by Target Outcome')
plt.ylabel('Units Approved (2nd Sem)')
plt.tight_layout()
plt.show()

# Plot 2: Boxplot - Curricular units 2nd sem (grade) by Target
plt.figure(figsize=(8, 5))
sns.boxplot(x='Target', y='Curricular units 2nd sem (grade)', data=df, palette='pastel')
plt.title('2nd Semester Grades by Target Outcome')
plt.ylabel('Grade Average (2nd Sem)')
plt.tight_layout()
plt.show()

# Plot 3: Boxplot - Admission Grade by Target
plt.figure(figsize=(8, 5))
sns.boxplot(x='Target', y='Admission grade', data=df, palette='Set3')
plt.title('Admission Grade by Target Outcome')
plt.ylabel('Admission Grade')
plt.tight_layout()
plt.show()
```

## 📊 Visual Insights from Boxplots

### 📦 1. 2nd Semester Approvals by Target

- 🎓 **Graduates** have the **highest number of approved subjects** in the second semester.
- ❌ **Dropouts** show the **lowest number of approvals**.
- 📉 There is a clear performance gap between Dropouts and Graduates, suggesting that **second-semester success is a strong indicator of student retention and graduation**.

📌 **Interpretation:** Students who do well in their second semester are far more likely to stay on track and complete their studies.

---

### 📘 2. 2nd Semester Grades by Target

- 🎓 Graduates maintain **higher average grades** in the second semester.
- ❌ Dropouts have **lower scores** and a tighter, more compressed grade distribution (less variance).
- 📉 Enrolled students typically fall in the middle.

📌 **Interpretation:** Not just passing, but **performing well** in second-semester courses is closely tied to graduation outcomes.

---

### 🎓 3. Admission Grade by Target

- 🧮 All three target groups (Dropout, Enrolled, Graduate) have **similar admission grades**.
- 🟨 Only a **slight increase** is seen among graduates.

📌 **Interpretation:** Although admission scores show some effect, their influence is **much weaker** than the academic performance during the program — especially second-semester progress.

---

✅ These visuals support the earlier correlation results and help highlight **what truly influences dropout risk**.


## 📝 Conclusion

- Students who pass and score well in their second semester are far more likely to graduate.
- Admission grades have only a weak link to long-term success.
- Dropouts are heavily concentrated among students with poor second-semester performance.
- These insights could help educational institutions design better student support in early semesters.
