## 🧾 Feature Descriptions – Wine Quality (White)

| Feature | Description | Role in Wine Quality |
|---------|-------------|-----------------------|
| **fixed acidity** | Non-volatile acids like tartaric acid that do not evaporate easily. | Provides structure and body to wine; excessive levels may make wine taste overly tart. |
| **volatile acidity** | Acetic acid (vinegar-like) that evaporates readily. | High levels produce unpleasant vinegar-like aromas, negatively affecting quality. |
| **citric acid** | Found in small amounts; adds freshness and flavor complexity. | Moderate influence; balanced levels improve taste and acidity perception. |
| **residual sugar** | Sugar left after fermentation (g/L). | Affects sweetness; higher values can be desirable or undesirable based on wine style. |
| **chlorides** | Amount of salt content in wine (mostly sodium chloride). | High levels may lead to salty or off-flavors; typically undesirable. |
| **free sulfur dioxide** | Form of SO₂ that prevents microbial growth and oxidation. | Acts as a preservative; too much can create chemical smells, too little risks spoilage. |
| **total sulfur dioxide** | Sum of free and bound SO₂. | Regulates wine stability; must be within acceptable limits for safe consumption. |
| **density** | Wine density, related to sugar and alcohol content. | Can indicate wine concentration; lower density usually means higher alcohol. |
| **pH** | Acidity level (inverse logarithmic scale). | Affects perception of freshness; ideal wines usually range from 2.9 to 3.5 pH. |
| **sulphates** | Additive contributing to SO₂ levels, also acts as an antioxidant. | Positively influences microbial stability and wine longevity. |
| **alcohol** | Percentage of alcohol by volume (% vol). | Generally correlates positively with wine quality; higher alcohol levels improve mouthfeel. |
| **quality** | Quality score (0–10) based on sensory data. | Target variable; represents the perceived quality of wine from human tasting panels. |


## 4. Descriptive Statistical Analysis


```python
# Measures of Central Tendency and Variability
mean_values = df.mean()
median_values = df.median()
mode_values = df.mode().iloc[0]
std_values = df.std()
var_values = df.var()

# Combine into a DataFrame
summary_stats = pd.DataFrame({
    'Mean': mean_values,
    'Median': median_values,
    'Mode': mode_values,
    'Std Dev': std_values,
    'Variance': var_values
})

summary_stats
```

```python
plt.figure(figsize=(12,5))
sns.barplot(x=summary_stats.index, y=summary_stats['Mean'])
plt.title('Mean Value of Each Feature')
plt.xticks(rotation=45)
plt.ylabel('Mean')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.barplot(x=summary_stats.index, y=summary_stats['Median'], color='orange')
plt.title('Median Value of Each Feature')
plt.xticks(rotation=45)
plt.ylabel('Median')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.barplot(x=summary_stats.index, y=summary_stats['Mode'], color='green')
plt.title('Mode Value of Each Feature')
plt.xticks(rotation=45)
plt.ylabel('Mode')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.barplot(x=summary_stats.index, y=summary_stats['Std Dev'], color='purple')
plt.title('Standard Deviation of Each Feature')
plt.xticks(rotation=45)
plt.ylabel('Standard Deviation')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.barplot(x=summary_stats.index, y=summary_stats['Variance'], color='red')
plt.title('Variance of Each Feature')
plt.xticks(rotation=45)
plt.ylabel('Variance')
plt.tight_layout()
plt.show()
```

```python
features = df.columns[:-1]  # All except 'quality'

for feature in features:
    plt.figure(figsize=(10, 4))
    sns.boxplot(x='quality', y=feature, data=df)
    plt.title(f'{feature} vs Wine Quality')
    plt.show()
```

### 🔍 Boxplot Interpretation Summary

- **Alcohol** and **volatile acidity** are the **most impactful** features influencing wine quality, showing clear upward and downward trends respectively.
- Features like **residual sugar**, **chlorides**, and **citric acid** showed **no significant change** across quality levels.
- **Outliers** exist in many features (e.g., residual sugar), suggesting some rare wine profiles.
- **Density**, **sulphates**, and **free sulfur dioxide** exhibit slight patterns but may not be strong independent predictors.

These insights help us focus on the **key drivers of wine quality**, which could later guide product improvement or classification.


```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='quality', y='alcohol', data=df, palette='Set2')
plt.title('Alcohol Distribution Across Wine Quality Levels')
plt.show()
```

## 🎻 Violin Plot Insights – Alcohol vs Wine Quality

The violin plot visualizes the **distribution of alcohol content** across different wine quality scores. Unlike boxplots, violin plots show the **density** of data points, revealing how values are concentrated.

### Key Observations:
- **Higher quality wines (scores 7, 8, 9)** tend to have **higher alcohol content** compared to lower quality wines.
- The distribution for **quality 5 and 6** is broader, indicating a **wider range of alcohol levels** among average-rated wines.
- **Lower quality wines (scores 3 and 4)** have **tighter and lower alcohol distributions**, suggesting they are consistently low in alcohol.
- The peak density shifts **upward** with quality — indicating that **alcohol is positively correlated** with wine quality.

### Conclusion:
> Alcohol content is a strong predictor of wine quality in this dataset. The violin plot helps reinforce the insight that **higher alcohol wines tend to be rated better** by human tasters. This supports our earlier correlation and boxplot analysis.


```python
# Only strong correlations with 'quality'
corr = df.corr()
quality_corr = corr['quality'].drop('quality').sort_values(ascending=False)

plt.figure(figsize=(8, 6))
sns.heatmap(df[quality_corr.index].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Among Top Features Influencing Quality')
plt.show()
```

## 🔥 Correlation Heatmap Insights – Wine Quality

We computed the **Pearson correlation coefficient** between all numerical features and the target variable `quality`. The strongest correlations (positive or negative) were selected and visualized in a focused heatmap.

### 📌 Top Features Correlated with `quality`:

| Feature              | Correlation with `quality` |
|----------------------|----------------------------|
| **alcohol**          | **+0.44**                  |
| **sulphates**        | +0.19                      |
| **citric acid**      | +0.16                      |
| **residual sugar**   | +0.01                      |
| **density**          | −0.31                      |
| **volatile acidity** | **−0.20**                  |
| **free sulfur dioxide** | −0.05                  |
| **total sulfur dioxide** | −0.17                |

---

### 💡 Key Insights:
- **Alcohol** shows the **strongest positive correlation** (+0.44) with wine quality, supporting visual trends from violin plots and boxplots.
- **Volatile acidity** shows a **clear negative correlation** (−0.20), implying that wines with less acetic acid tend to score better.
- **Density** (−0.31) is also negatively correlated — likely due to its inverse relation with alcohol and sugar concentration.
- **Sulphates** and **citric acid** show **weak positive correlations**, suggesting a mild influence on perceived quality.
- **Residual sugar**, **free SO₂**, and **total SO₂** show near-zero to weak negative correlation, indicating **little direct impact** on quality ratings.

---

### ✅ Conclusion:
> The heatmap numerically confirms that **alcohol content is the most important feature** influencing wine quality, while **volatile acidity and density negatively affect** the perception of quality. These findings align with earlier visual explorations and reinforce the role of specific chemical traits in wine assessment.
